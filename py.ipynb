```python
def convert(x):
    temp = x.split('-')
    if len(temp) == 2:
        return (float(temp[0]) + float(temp[1])) / 2
    try:
        return float(x)
    except:
        return None
```

```python
def removeoutliers(df):
    df_output = pd.DataFrame()
    for key,subdf in df.groupby('location'):
        m = np.mean(subdf.price_per_feet)
        std = np.std(subdf.price_per_feet)
        gendf = subdf[(subdf['price_per_feet'] > (m - std)) & (subdf.price_per_feet <=  (m + std))]
        df_output = pd.concat([df_output, gendf],ignore_index=True)
    return df_output
data = removeoutliers(data)
data.describe()
```

```python
def bhk_outlier_remove(df):
    exclude_indices = np.array([])
    for location, location_df in df.groupby('location'):
        bhk_stats = {}
        for bhk, bhk_df in location_df.groupby('bhk'):
            bhk_stats[bhk] = {
                'mean' : np.mean(bhk_df.price_per_feet),
                'std' : np.std(bhk_df.price_per_feet),
                'count':bhk_df.shape[0]
            }
        print(location, bhk_stats)
        for bhk, bhk_df in location_df.groupby('bhk'):
            stats = bhk_stats.get(bhk - 1)
            if stats and stats['count']>5:
                exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_feet < (stats['mean'])].index.values)
    return df.drop(exclude_indices,axis='index')
```

```python
data = bhk_outlier_remove(data)
```

```python
x = data.drop(columns=['price'])
y = data['price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.preprocessing import StandardScaler, OneHotEncoder

from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
print(x_train.shape)
print(x_test.shape)
```

Applying the linear Regression

```python
column_transformer = make_column_transformer((OneHotEncoder(sparse_output=False), ['location']),remainder='passthrough')
```

```python
scaler = StandardScaler()
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(column_transformer, scaler, lr)
```

```python
pipe.fit(x_train, y_train)
```

```python
y_pred_lr = pipe.predict(x_test)
```

```python
r2_score(y_test, y_pred_lr)
```

Apply Lasso

```python
lasso = Lasso()
```

```python
pipe = make_pipeline(column_transformer, scaler, lasso)
```

```python
pipe.fit(x_train, y_train)
```

```python
ypred_lasso = pipe.predict(x_test)
r2_score(y_test, ypred_lasso)
```

Apply Ridge

```python
ridge = Ridge()
```

```python
pipe = make_pipeline(column_transformer, scaler, ridge)
```

```python
pipe.fit(x_train, y_train)
```

```python
ypred_ridge = pipe.predict(x_test)
r2_score(y_test, ypred_ridge)
```

```python
print("no regulization: ", r2_score(y_test, y_pred_lr))
print("lasso: ", r2_score(y_test, ypred_lasso))
print("ridge: ", r2_score(y_test, ypred_ridge))
```
